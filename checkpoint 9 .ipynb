```python
dataset=[['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
dataset
```

```python
import pandas as pd
df=pd.DataFrame(data=dataset)
df
```

```python
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

### Apriori algorithm.

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

nous pouvons voir que the support de pants, scarf et de pants, skirts est de 0.6 . cette boutique de vetement doit alore mettre les pantalons et les jupes a coté les unes des autres , et les assortir ensemble . de meme pour les pantalons et les echarpes . en ce qui concerne les autres produits de ce magasin , et vu que leurs fréquence d'achat est minime , le gestionnaire de cette boutique doit faire des sondages ou demander aux consommateurs pourquoi n'achètent-ils pas ces produits afin d'essayer de les améliorer pour que les consommateurs puissent les acheter  

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

cela montre que les gens préfèrent acheter des pantalons et des jupes ensemble qu'acheter soit un pantalon soit une jupe . ces deux produis devraient absolument etre mis ensemble

```python
import pandas as pd
df=pd.read_csv('Market_Basket_Optimisation.csv')
df
```

```python
import numpy as np
df= df.replace(np.nan, None, regex=True)
df
```

```python
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(df).transform(df)    #Apply one-hot-encoding on our dataset
dff=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
dff
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```
